```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

Load the summary csv. 

```python
df = pd.read_csv('Result_Summary.csv')
# removed un-named columns
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
assert df.shape == (445, 10) # we inclulde 445 questions
```

Plot the summary

```python
category_count = df['Category'].value_counts().sort_values(ascending=True)
category_percent = category_count / category_count.sum() * 100
ax = category_count.plot(kind='barh', title='Categories of NCLEX Practice Questions')
ax.title.set_fontsize(10)
ax.set_xlabel('Quantity', fontsize=8)
ax.set_ylabel('')

# set background color
# ax.set_facecolor('#BCBCBC')

# set color inside the bar
bars = ax.patches
for i in range(len(bars)):
    bars[i].set_facecolor('#CBE7EF')

# set spines
ax.spines['top'].set_visible(True)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(True)

# set backendgorund horizontal line
ax.yaxis.grid(True, color='#EEEEEE')
ax.xaxis.grid(False)
ax.set_axisbelow(True)

# print pentage and quantity on the right side of the bar
for i in range(len(category_count)):
    plt.text(category_count.iloc[i], i, f'{category_count.iloc[i]} ({category_percent.iloc[i]:.2f}%)', ha='left', va='center', fontsize=8, color='#084251')

# set the font size of the category name (bar name)
plt.yticks(fontsize=8, color='#084251')

# change the dpi of the figure
plt.rcParams['figure.dpi'] = 600
# change the size of the figure : 7.24 × 6.28 inches
plt.rcParams['figure.figsize'] = [7.24, 6.28]

plt.show()
```
